### 丟一個銅板，丟了100次，出現正面 50 次的機率有多大。
(提示: 先想是哪一種分配，然後透過 python 語法進行計算)


```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
p = 0.5 #probability of flipping a fair coin
n = 100  # flip 100 times
r = np.arange(0,101) #possible number of heads
#print(type(k))

probs = stats.binom.pmf(r, n, p)

#probability of 50 heads
print(probs[50])
```

```python
#check the distribution of binomial PMF
plt.bar(r, probs)
plt.ylabel('P(X=x)')
plt.xlabel('x')
plt.title('binomial(n=100,p=0.5)')
plt.show()
```

```python
# 載入需要的...
import matplotlib.pyplot as plt
import numpy as np
```

```python
#畫出完整的 cos 圖形
x = np.arange(0, 3 * np.pi, 0.1)
y_cos = np.cos(x)

plt.plot(x, y_cos)
plt.show()
```

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

#作業
#方法不只有一種

T = np.arctan2(Y, X)
plt.scatter(X, Y, alpha = 0.5, c = T, s = 50)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.show()
```

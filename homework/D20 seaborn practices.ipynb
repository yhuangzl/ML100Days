```python
import numpy as np
import seaborn as sns
```

```python
#設定圖形樣式 - whitegrid
# use sns.set
sns.set(style = 'whitegrid')

# 利用 NUMPY 去建立資料集
# np.random.RandomState 設定數學式
rs = np.random.RandomState(0)
x = rs.normal(2, 1, 75)
y = 2 + 1.5 * x + rs.normal(0, 2, 75)

# 畫圖
# sns.residplot
sns.residplot(x = x, y = y)
```

```python
# bin: 指的是特徵值, 
# kde: on/off
# sns.distplot();
sns.distplot(x, kde=False, bins = 75, rug = True)
```

```python
sns.displot(x, rug = True, kde = True, bins = 15)
```

### Q2: 運用範例的 python 程式碼，計算大樂透的中頭獎機率?

```python
# library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
```

```python
N= 49 #total 49 numbers
K= 6 #6 lucky numbers of the 1st price
n= 6 #6 numbers for 1 bet

#產生一個同樣間隔的序列
r = np.arange(0, min(n+1,K+1)) #產出 x 對應點

# 2.計算機率質量函數 (probability mass function)
# P(X=x) --> 是機率
probs = stats.hypergeom.pmf(r, N,K,n)
print('The prob of winning the 1st price =', probs[6])

plt.bar(r, probs)
plt.ylabel('P(X=x)')
plt.xlabel('numbers of hit')
plt.title('pmf of Hypergeometric(N=49,K=6,n=6)')
plt.show()
```
